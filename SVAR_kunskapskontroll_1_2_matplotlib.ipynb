### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# The Object Oriented approach or the Pyplot/Matplot approach. The OO-approach is the recommended one, atleast for bigger projects.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure is the canvas where everything goes on.
# Axes are the objects on the canvas i.e Titles, data points, labels etc
# Axis are the actually x,y and sometimes z axis. You can set different tick and ticklabels depending on the measurements you want.
# Artist is basically everything you render on the Figure, but you could render another figure onto the figure therefor even a figure is an Artist.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# The expected input data type is numpy.array or objects that can be converted to numpy.asarray
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Object Oriented Plot

data = np.linspace(-4, 4, 100)

fig, ax = plt.subplots(figsize=(4,4))
ax.plot(data, x**2, label='Squared')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Object Oriented Plot')
ax.legend()

# Pyplot

plt.figure(figsize=(4,4))
plt.plot(data, x**2, label='Squared')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Pyplot')
plt.legend()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Scatter plot

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
ax1.scatter(random_data_x, random_data_y)
ax1.plot(x, y, color = 'red')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Bar plot
ax2.bar(names, values)
ax2.set_xlabel('Fruits')
ax2.set_ylabel('Quantity')
ax2.set_title('Bar Plot')
```
